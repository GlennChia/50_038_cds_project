```python
import pandas as pd
```

```python
DATA_DIR = "../../data/raw/"
INPUT_FILE_NAME = "TED_Talks_by_ID_plus-transcripts-and-LIWC-and-MFT-plus-views.csv"
OUTPUT_FILE_NAME = 'subset_raw.parquet'
```

```python
df = pd.read_csv(DATA_DIR + INPUT_FILE_NAME)
df.head()
```

```python
#list(df.columns)
```

```python
cols_to_keep = ['speaker', 'headline', 'description', 'duration', 'tags', 'transcript', 'WC']
subset_df = df.loc[:, cols_to_keep]
subset_df.head()
```

```python
subset_df.to_parquet(DATA_DIR + OUTPUT_FILE_NAME)
```
